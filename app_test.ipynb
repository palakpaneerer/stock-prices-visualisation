```python
#一連の流れのコードを纏める。

#企業名、ピリオド指定
aapl = yf.Ticker('AAPL')
days = 20

#株価情報取得
hist = aapl.history(period=f'{days}d')

#インデックスの時間表示方法の変更
hist.index = hist.index.strftime('%d %B %Y')

#Closeデータのみ抽出
hist = hist[['Close']]

#コラム名を企業名に設定
hist.columns = ['apple']

#企業名をインデックスに表示したいため、転置
hist = hist.T

#インデックス名をnameに設定
hist.index.name = 'Name'

#最終確認
hist
```

```python
#上記のコードを汎用的にする。

#ピリオド、企業名を指定
days = 20
tickers = {
    'apple': 'AAPL',
    'meta': 'META',
    'google': 'GOOGL',
    'microsoft': 'MSFT',
    'netflix': 'NFLX',
    'amazon': 'AMZN'
}


#データを格納していく空のデータフレームを作成
df = pd.DataFrame()

#for文で1企業ずつ株価データを取得する。
for company in tickers.keys():
    #株価データ取得
    tkr = yf.Ticker(tickers[company])
    
    #ピリオド指定
    hist = tkr.history(period=f'{days}d')
    
    #日付データの表示方法変更
    hist.index = hist.index.strftime('%d %B %Y')
    
    #Closeデータのみ抽出
    hist = hist[['Close']]
    
    #コラム名を企業名に変更
    hist.columns = [company]

    #インデックスに企業名が来るように転置
    hist = hist.T

    #インデックス名をNameに変更
    hist.index.name = 'Name'

    #dfにデータを結合、加えていく。
    df = pd.concat([df, hist])

#最終確認
df
```

```python
#関数化

def get_data(days, tickers):
    df = pd.DataFrame()
#for文で1企業ずつ株価データを取得する。
    for company in tickers.keys():
        #株価データ取得
        tkr = yf.Ticker(tickers[company])
        #ピリオド指定
        hist = tkr.history(period=f'{days}d')
        #日付データの表示方法変更
        hist.index = hist.index.strftime('%d %B %Y')
        #Closeデータのみ抽出
        hist = hist[['Close']]
        #コラム名を企業名に変更
        hist.columns = [company]
        #インデックスに企業名が来るように転置
        hist = hist.T
        #インデックス名をNameに変更
        hist.index.name = 'Name'
        #dfにデータを結合、加えていく。
        df = pd.concat([df, hist])
    return df
```

```python
#関数実行

#関数に必要な変数設定
days = 20
tickers = {
    'apple': 'AAPL',
    'meta': 'META',
    'google': 'GOOGL',
    'microsoft': 'MSFT',
    'netflix': 'NFLX',
    'amazon': 'AMZN'
}

#関数実行
df = get_data(days, tickers)

#最終確認
df
```
